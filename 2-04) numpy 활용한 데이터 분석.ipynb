# 문제) 각 사용자 별 평균 평점 계산하기

```python
# user에 대해 unique 값만 먼저 뽑기
user_id = np.unique(data[:,0])
```

```python
user_id[:10]
```

```python
user_id.shape
```

```python
# 빈 리스트 생성
mean_rating_by_user_list = []

for user in user_id:
    data_for_user = data[data[:,0] == user, :]           # 특정 user_id 값들의 array를 저장한 변수 (unique 값별로 저장)
    mean_rating_for_user = data_for_user[:, 2].mean()    # 특정 user_id의 평점을 평균
    mean_rating_by_user_list.append([user, mean_rating_for_user]) # 빈 리스트에 결과 넣기
```

```python
data_for_user[:10]
```

```python
mean_rating_by_user_list[:10]
```

# 결과를 파일에 writing하여 저장하기

```python
# 리스트 -> array 타입으로 변환
mean_rating_by_user_list_array = np.array(mean_rating_by_user_list, dtype=np.float64)
mean_rating_by_user_list_array.shape
```

```python
#파일에 저장
np.savetxt("./data/mean_rating_by_user.csv", mean_rating_by_user_list_array, fmt="%.3f", delimiter=',')
```
